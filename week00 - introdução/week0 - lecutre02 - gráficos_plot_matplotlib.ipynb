### seno de cada valor x

```python
# lista vazia ysin, que será preenchida com .append()
ysin = []

for x in xsin:
    ysin.append(math.sin(x))
    
#ysin
```

```python
plt.plot(xsin, ysin, color='green', linewidth=2)
plt.title('senóide')
plt.xlabel("seq x")
plt.ylabel("seno(x)");
```
